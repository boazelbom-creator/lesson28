```python
import numpy as np
```

```python
x=np.array([-40,-10,0,8,15,22,38])
```

```python
y=np.array([-40,14,32,46.4,59,71.6,100])
```

```python
import keras as kr
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Create a Sequential model
model = Sequential()

# Add a Dense layer with 1 neuron and specify the input shape
model.add(Dense(units=1, input_shape=[1]))

# Print the model summary to see the layer configuration
model.summary()
```

```python
from keras.optimizers import Adam

# Compile the model with mean squared error loss and Adam optimizer
model.compile(loss='mean_squared_error', optimizer=Adam(learning_rate=0.1))
```

```python
# Train the model using the defined x and y vectors
# The batch_size is set to the number of observations (7) as requested
history = model.fit(x, y, epochs=500, batch_size=7)
```

```python
import matplotlib.pyplot as plt

# Get the loss values from the history object
loss = history.history['loss']

# Create a range for the number of epochs
epochs = range(len(loss))

# Plot the loss against the epochs
plt.plot(epochs, loss, 'r', label='Training loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.legend()
plt.show()
```

```python
predicted_y = model.predict(np.array([100.0]))
print(f"The predicted value of y when x is 100 is: {predicted_y[0][0]:.2f}")
```

```python
# Get the weights and biases from the trained model
weights = model.get_weights()
m = weights[0][0][0] # Weight (slope)
b = weights[1][0]    # Bias (y-intercept)

# Print the equation of the line
print(f"The learned linear equation is: y = {m:.2f}x + {b:.2f}")
```
